```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

```

```python
# Define os limites dos anos realistas para carros
limite_inferior = 1990
limite_superior = 2020

# Filtra os dados para incluir apenas anos dentro dos limites
df_cars_filtrado = df_cars[(df_cars['powerPS'] >= limite_inferior) & (df_cars['powerPS'] <= limite_superior)]

# Plota o histograma com as bordas pretas nas barras
plt.figure(figsize=(12, 6))
plt.hist(df_cars_filtrado['powerPS'], bins=len(df_cars_filtrado['powerPS'].unique()), alpha=0.5, edgecolor='black')
plt.xlabel('Ano de FabricaÃ§Ã£o')
plt.ylabel('FrequÃªncia')
plt.title('Histograma de Ano de FabricaÃ§Ã£o')
plt.show()
```

```python
import plotly.express as px

# Limitar os dados de 1960 a 2020
df_cars_filtered = df_cars[(df_cars['yearOfRegistration'] >= 1965) & (df_cars['yearOfRegistration'] <= 2018)]

# Plotar o histograma com os dados filtrados
fig = px.histogram(data_frame=df_cars_filtered, x="yearOfRegistration", y="price", range_x=[1965, 2018], barmode='overlay',
                   title='Histograma de Ano de FabricaÃ§Ã£o e PreÃ§o',
                   labels={'yearOfRegistration': 'Ano de FabricaÃ§Ã£o', 'kilometer': 'Kilometro'},
                   opacity=0.75,
                   histnorm='percent',
                   marginal='rug',
                   color_discrete_sequence=['blue'])
fig.update_traces(marker=dict(line=dict(color='black', width=1)))
fig.show()
```
